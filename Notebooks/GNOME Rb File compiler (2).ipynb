```python
import pandas as pd
import os

path = '\\Users\\su6676\\GNOME\\6-24-6-25'
os.chdir(path)

# File paths
file1_path = 'TEK00060.csv'
file2_path = 'TEK00061.csv'
file3_path = 'TEK00062.csv'
output_file_path = 'output.csv'

# Convert to panda data frame
file1_data = pd.read_csv(file1_path)
file2_data = pd.read_csv(file2_path)
file3_data = pd.read_csv(file3_path)

# Combine files

df_file = pd.concat([file1_data,file2_data,file3_data], axis=1, ignore_index=False)

# Remove all the garbage the Oscilliscope spits out
corrected_df = df_file.iloc[13:].reset_index(drop=True)
 

# Drop duplicate rows
df = corrected_df.T.drop_duplicates().T

# Save the modified data to a new CSV file
df.to_csv(output_file_path, index=False, header=False)

print(f"Modified data saved to {output_file_path}")
```
